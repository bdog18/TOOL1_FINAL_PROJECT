```python
import pandas as pd
```

Data Loading

```python
terror_df = pd.read_csv("GlobalTerrorism/GlobalTerrorismDataset.csv", encoding="latin-1")

life_df = pd.read_csv("LifeExpectancy/LifeExpectancyAtBirth.csv", encoding="latin-1")
```

Data Cleaning

```python
# Get min and max years for both dfs
life_yr_min = life_df["Year"].min()
life_yr_max = life_df["Year"].max()
terror_yr_min = terror_df["iyear"].min()
terror_yr_max = terror_df["iyear"].max()

# Filter dfs so that they are both using the same year ranges
terror_df = terror_df[(terror_df["iyear"] >= life_yr_min) & (terror_df["iyear"] <= life_yr_max)]
life_df = life_df[(life_df["Year"] >= terror_yr_min) & (life_df["Year"] <= terror_yr_max)]
```
